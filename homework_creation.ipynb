```python
def chek_homework(llm,tasks,teach_solution,stud_solution,points):
        template_chek_homework = """you are a teacher's assistant,your task is to check the student's
                                    homework based on the example of the teacher's solution and give points for each task,
                                     taking into account that for each task the student can get a maximum of the following
                                    If everything is done correctly, you should output only the word "correct",
                                    if there are errors in the solution, you should describe the error as briefly as possible,
                                    for example: the problem "there are two apples, one was taken away, how many are left?"
                                    the student's answer is "2-1=0"
                                    your conclusion "incorrect subtraction of numbers"
                                    The output should be only in the format and don`t add any disciption for your output:
                                      ['task 1': ["correct" , 4],
                                       'task 2': ["incorrect subtraction of numbers" , 3],
                                       'task 3': ["misunderstanding of the task" , 1]]
                                    task:
                                    {task}
                                    teacher's solution:
                                    {teach_solution}
                                    maximum points of task
                                    {points}
                                    student's solution:
                                    {stud_solution}
                                   """
        prompt_chek_homework = PromptTemplate(
            input_variables=["tasks", "teach_solution","stud_solution", "points"],
            template=template_chek_homework,
            output_variables=["output"],
            verbose=True,
        )
        get_chek_homework = LLMChain(
            llm=llm, prompt=prompt_chek_homework, output_key="output", verbose=True
        )

        proven_homework = get_chek_homework(
                                    {
                                        "tasks": tasks,
                                        "teach_solution": teach_solution,
                                        "stud_solution": stud_solution,
                                        "points": points,
                                    }
                                )
        return proven_homework['output']
```

```python
def do_homework(llm,tasks):
        template_chek_homework = """you need to do your
                                  homework as if you were a 6th grader some of
                                   the tasks you have to do incorrect- it is important
                                    {tasks}
                                   """
        prompt_chek_homework = PromptTemplate(
            input_variables=["tasks"],
            template=template_chek_homework,
            output_variables=["output"],
            verbose=True,
        )
        get_chek_homework = LLMChain(
            llm=llm, prompt=prompt_chek_homework, output_key="output", verbose=True
        )

        homework = get_chek_homework(
                                    {
                                        "tasks": tasks,

                                    }
                                )
        return homework['output']
```

```python
def generate_text(model, text, hobby):
    prompt = (
        "Input: A math problem text and a hobby of a kid.\n"
        "Task: Reformat the text of the math problem to incorporate the kid's hobby "
        "without altering the context of math problem or numerical values."
        f"Math Problem: \"{text}\"\n"
        f"Hobby: {hobby}\n"
        "Reformatted Problem:"
    )
    openai.api_key = "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"
    client = openai.ChatCompletion()
    response = client.create(
        model=model,
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": text}
        ])

    return response.choices[0].message['content']

def reformat_text(text, hobbies):
    tasks = []
    model = "gpt-4-1106-preview"

    for hobby in hobbies:
        reform_text = generate_text(model, text, hobby)
        tasks.append(reform_text)

    return tasks
```

```python
def get_dict_from_proven_homework (proven_homework) :
      proven_homework = "{" + proven_homework[1:-1] + "}"
      proven_homework = eval(proven_homework)
      return proven_homework
```
